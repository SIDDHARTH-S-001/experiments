Principle Component Analysis

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Step 1: Load the image
img_path = 'D:\\Onedrive\\experiments\\experiments\\ML\\pre_processing\\man_face.jpg'
image = cv2.imread(img_path)

# Step 2: Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Reshape the 2D image into a 2D array for PCA
# Each row is a sample, each column is a feature
h, w = gray_image.shape
reshaped_image = gray_image.reshape(h, w)

# Step 4: Apply PCA to the reshaped 2D image
# We set n_components=3 to get the first 3 principal components
pca = PCA(n_components=4)
transformed_image = pca.fit_transform(reshaped_image)

# Step 5: Inverse transform to reconstruct the image from its first 3 principal components
reconstructed_image = pca.inverse_transform(transformed_image)

# Step 6: Visualize the original image and its principal components
plt.figure(figsize=(10, 10))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# First Principal Component Reconstruction
plt.subplot(2, 2, 2)
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Image after PCA Reconstruction (3 Components)')
plt.axis('off')

plt.tight_layout()
plt.show()

# To view variance explained by each component:
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by the 1st, 2nd, and 3rd components: {explained_variance}")
```
